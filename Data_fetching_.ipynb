```python
import pandas as pd
import numpy as np
import requests
import json as js
from time import sleep
```

```python
# read given dataset id and dialect columns
data_ID_Label = pd.read_csv('/content/dialict.csv',dtype='string')
data_ID_Label
```

```python
# split dataset into N of lists with size 1000 to be suitable with post request requirement  

# add ID into list
data_ID_Label_list = list(data_ID_Label.id)
data_chunked_list = list()
# set size of each list
chunk_size = 1000

for i in range(0, len(data_ID_Label_list), chunk_size):
    data_chunked_list.append(data_ID_Label_list[i:i+chunk_size])
```

```python
# define url and requrest list
round = 0
count = 0
while round < len(data_chunked_list):

  url = 'https://recruitment.aimtechnologies.co/ai-tasks'
  request = js.dumps(data_chunked_list[round])

  # send request
  response = requests.post(url, data = request)
  # sleep(1)

  # read response as json and convert to pd.dataFrame
  response_json = response.json()
  response_dataframe = pd.DataFrame(list(response_json.items()),
                        columns=['tweet_id','tweet'])

  round += 1
  count += len(response_dataframe)

  print(f"Done request no.{round} and get {count} tweets ")

 # save and append output into csv file
  with open(f'output.csv', mode='a') as f:
    response_dataframe.to_csv(f, mode ='a', header=f.tell()==0)



```
